```python
import meilisearch
import os
from dotenv import load_dotenv

load_dotenv()  # .env 파일 로드

MASTER_key = os.getenv("MAILSEARCH_API_MASTER_KEY")
MAILSEARCH_LOCAL_HOST_t = os.getenv("MAILSEARCH_LOCAL_HOST")
```

```python
client = meilisearch.Client(MAILSEARCH_LOCAL_HOST_t, MASTER_key)
```

```python
client
```

```python
import pandas as pd
df = pd.read_csv('nasdaq_screener_1755841220627.csv', na_filter=False)
df.head(4)
```

```python
df['id'] = df['Symbol'].str.strip().replace(r'[/^]', '_', regex=True)
df.head(3)
```

```python
client.index('nasdaq').add_documents(df.to_dict(orient='records'), primary_key='id')
```

```python
client.index('nasdaq').search('appl')
```
